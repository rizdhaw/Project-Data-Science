### Training

```python
def create_models(seed=2):
    models = []
    models.append(('dummy_classifier', DummyClassifier(random_state=seed, strategy='most_frequent')))
    models.append(('k_nearest_neighbors', KNeighborsClassifier()))
    models.append(('logistic_regression', LogisticRegression(random_state=seed)))
    models.append(('support_vector_machines', SVC(random_state=seed)))
    models.append(('random_forest', RandomForestClassifier(random_state=seed)))
    models.append(('gradient_boosting', XGBClassifier(random_state=seed)))
    
    return models

# create a list with all the algorithms we are going to assess
models = create_models()
```

```python
# test the roc_auc of each model using default hyperparameters

results = defaultdict(list)

scoring = 'roc_auc'
for name, model in models:
    # fit the model with the training data
    model.fit(X_train, y_train)

    # make predictions with the training data
    predictions_train = model.predict(X_train)
    
    # make predictions with the testing data
    predictions_test = model.predict(X_test)
    
    # calculate roc_auc 
    roc_auc_train = roc_auc_score(y_train, predictions_train)
    
    # calculate roc_auc 
    roc_auc_test = roc_auc_score(y_test, predictions_test)
    
    # append the model name and the roc_auc to the lists
    results['name'].append(name)
    results['ROC-AUC train'].append(roc_auc_train)
    results['ROC-AUC test'].append(roc_auc_test)

display(pd.DataFrame(results)) 
```

## Model Evaluation

```python
# define kfold paramaters

def model_evaluate(model, X_train, y_train, X_test, y_test):
    cv = KFold(n_splits=10, random_state=42, shuffle=True)
    scores = cross_val_score(model, X_train, y_train, scoring='roc_auc', cv=cv, n_jobs=-1)
    scores = np.abs(scores)
    print('Train Kfold Mean ROC-AUC', scores.mean())

    model.fit(X_train, y_train)
    y_test_predict = model.predict(X_test)
    print('ROC-AUC Test', roc_auc_score(y_test, y_test_predict))
    print('Accuracy Test', accuracy_score(y_test, y_test_predict))

    # creating ROC curves
    probabilities_valid = model.predict_proba(X_test)
    probabilities_one_valid = probabilities_valid[:, 1]
    fpr, tpr, thresholds = roc_curve(y_test, probabilities_one_valid)
    auc_roc = roc_auc_score(y_test, probabilities_one_valid)

    print()

    plt.figure(figsize=(3, 3))
    plt.plot(fpr, tpr)
    plt.plot([0, 1], [0, 1], linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.show()

    return
```
