## Mempelajari dan Menangani Outlier


```python
# Tentukan batas bawah outlier 
def batas_bawah(data_column):
    Q1 = data_column.quantile(0.25)
    Q3 = data_column.quantile(0.75)
    IQR = Q3 - Q1
    batas_bawah = Q1 - (1.5 * IQR)
    return batas_bawah
```

```python
# Tentukan batas atas outlier 
def batas_atas(data_column):
    Q1 = data_column.quantile(0.25)
    Q3 = data_column.quantile(0.75)
    IQR = Q3 - Q1
    batas_atas = Q3 + (1.5 * IQR)
    return batas_atas
```

```python
# Simpan data tanpa outlier dalam DataFrame yang terpisah
# batas atas dan bawah outlier car_age
df_filter = df[(df['car_age'] >= batas_bawah(df['car_age']))&(df['car_age'] <= batas_atas(df['car_age']))]
# batas atas dan bawah outlier price
df_filter = df_filter[(df_filter['price'] >= batas_bawah(df_filter['price']))&(df_filter['price'] <= batas_atas(df_filter['price']))]
# batas atas dan bawah outlier odometer
df_filter = df_filter[(df_filter['odometer'] >=batas_bawah(df_filter['odometer']))&(df_filter['odometer'] <= batas_atas(df_filter['odometer']))]
#df_filter = df_filter.query('odometer != 0')
```

- Tipe kendaraan besar seperti bus, truck, pickup, dan offroad memiliki rata-rata harga sekitar 15,000 hingga 17,500 dollar. 
- Tipe kendaraan mewah coupe dan convertible memiliki rata-rata harga sekitar 12,500 - 13,000 dollar 
- Tipe kendaraan penumpang dengah rata-rata harga sekitar 6000 - 10500 dollar

Selain tipe, faktor kategorikal yang cukup berpengaruh terhadap harga adalah warna. Kendaraan dengan warna custom, putih dan hitam relatif lebih tinggi dibandingkan warna lainnya. Mungkin karena warna custom memerlukan biaya ekstra karena sesuai dengan request buyer, sedangkan warna putih dan hitam secara selera cukup netral sehingga relatif dapat diterima oleh banyak kalangan. Warna yang tidak umum seperti orange memiliki harga yang relatif tinggi pada tipe mobil SUV. Namun jumlahnya tidak banyak dan tidak terjadi pada sedan.

Transmisi kendaraan juga dapat menjadi penentu harga kendaraan. Dimana mobil dengan transmisi automatic harganya lebih tinggi daripada transmisi manual
